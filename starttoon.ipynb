```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
import csv
import pandas as pd

dataset1 = pd.read_csv('Data_1.csv')
dataset2 = pd.read_csv('Data_2.csv')
plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.plot(dataset1, label='Dataset 1', color='b')
plt.title('Dataset 1')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(dataset2, label='Dataset 2', color='r')
plt.title('Dataset 2')
plt.xlabel('Index')
plt.ylabel('Value')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

mean1 = np.mean(dataset1)
median1 = np.median(dataset1)
std_dev1 = np.std(dataset1)
range1 = np.ptp(dataset1)

mean2 = np.mean(dataset2)
median2 = np.median(dataset2)
std_dev2 = np.std(dataset2)
range2 = np.ptp(dataset2)

mean1, median1, std_dev1, range1, mean2, median2, std_dev2, range2

print(f"Dataset 1 - Mean: {mean1}, Median: {median1}, Std Dev: {std_dev1}, Range: {range1}")
print(f"Dataset 2 - Mean: {mean2}, Median: {median2}, Std Dev: {std_dev2}, Range: {range2}")

```
